```python


from pyspark import SparkConf
from pyspark.sql import SparkSession

# Configurer Spark pour augmenter la mémoire des exécuteurs
conf = SparkConf().set("spark.executor.memory", "4g") \
                  .set("spark.driver.memory", "4g") \
                  .set("spark.memory.offHeap.enabled", "true") \
                  .set("spark.memory.offHeap.size", "4g")

spark = SparkSession.builder \
                    .config(conf=conf) \
                    .appName("Increase Memory Example") \
                    .getOrCreate()

# Lire le fichier Parquet
df = spark.read.parquet('data/filtered_df_output.parquet')

# Afficher les 5 premières lignes du DataFrame
df.show(5)


# Si vous souhaitez afficher le schéma du DataFrame
df.printSchema()
```

```python
import sqlite3

# Créer une connexion SQLite en mémoire
conn = sqlite3.connect(':memory:')

# Charger le DataFrame dans la base de données SQLite
df.to_sql('table_temp', conn, index=False, if_exists='replace')

# Créer un curseur pour exécuter des requêtes SQL
cursor = conn.cursor()

# Exécuter une requête SQL pour sélectionner des données
query = "SELECT * FROM 'table_temp' WHERE user_id = 568782581;"

cursor.execute(query)

# Récupérer les résultats
results = cursor.fetchall()

# Afficher les résultats
for row in results:
    print(row)

# Fermer la connexion
conn.close()
```

```python


# Filtrer les données pour afficher la ligne correspondant à user_id=568782581
user_data = df[df['user_id'] == 568782581]

# Afficher la ligne
print(user_data)
```
