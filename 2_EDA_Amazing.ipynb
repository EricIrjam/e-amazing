# 4. Analyse des corrélations

Matrice de corrélation pour voir comment les variables numériques sont liées entre elles.

```python

import matplotlib.pyplot as plt
import seaborn as sns


# Sélectionner uniquement les colonnes numériques pour la corrélation
numeric_cols = df.select_dtypes(include=['int64', 'float64', 'int32'])

# Calculer la matrice de corrélation
corr_matrix = numeric_cols.corr()

# Afficher la matrice de corrélation en utilisant un heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Variables')
plt.show()

```



### Points Clés de la Matrice de Corrélation :

Votre analyse est globalement correcte, mais il y a quelques ajustements à apporter pour affiner l'interprétation des corrélations observées :

1. **Corrélations Fortes (proches de 1 ou -1) :**
   - **`number_of_views_2m`, `number_of_views_5m`, `number_of_views_7m` :** Ces variables montrent des corrélations fortes entre elles (supérieures à 0.8), ce qui est attendu car elles représentent le même type de comportement (nombre de vues) sur différentes périodes. Cela suggère que les clients qui sont actifs en termes de vues tendent à maintenir ce comportement au fil du temps.
   
   - **`number_of_sessions_2m`, `number_of_sessions_5m`, `number_of_sessions_7m` :** De même, le nombre de sessions sur différentes périodes est fortement corrélé (corrélations supérieures à 0.7), ce qui indique une constance dans le niveau d’engagement des clients sur le site au fil des mois.

   - **`count_products_2m`, `count_products_5m`, `count_products_7m` :** Les corrélations élevées entre ces variables (autour de 0.9) montrent que les clients qui consultent plus de produits sur une période sont susceptibles de continuer ce comportement sur les autres périodes. Cela reflète une continuité dans les habitudes de navigation.

   - **`total_purchase_value` et `count_products` :** Bien que `total_purchase_value` montre une corrélation modérée à forte avec `count_products` sur plusieurs périodes (entre 0.4 et 0.5), ce n’est pas une corrélation extrêmement forte (proche de 1). 
   Cela suggère que même si les clients qui regardent plus de produits ont tendance à dépenser davantage, **cette relation n'est pas strictement linéaire et pourrait être modifiée par d'autres éléments tels que le prix des articles ou les habitudes d'achat**.

## 5. Analyse de la distribution catégorielle
Ici, nous examinons la répartition des catégories :

```python
# Fonction pour analyser une variable catégorielle
def analyze_categorical(df, column):
    # Affichage des valeurs uniques et de leur fréquence
    print(f"Analyse de la variable : {column}")
    print(df[column].value_counts())
    print("\n")
    
    # Visualisation de la distribution des catégories
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution de {column}")
    plt.xlabel(column)
    plt.ylabel("Nombre d'observations")
    plt.show()
```

```python


# Liste des colonnes catégorielles à analyser
categorical_columns = ['preferred_brand', 'preferred_category', 'most_active_time', 'most_active_day']

```

## 6. Analyse des outliers

```python
# Liste des colonnes à supprimer
columns_to_drop = ['number_of_views_2m', 'number_of_views_5m', 'number_of_carts_2m',
                   'number_of_carts_5m', 'number_of_sessions_2m', 'number_of_sessions_5m',
                   'count_products_2m', 'count_products_5m', 'avg_price_2m', 'avg_price_5m']

# Suppression des colonnes si elles existent dans le DataFrame
numeric_df = numeric_cols.drop(columns=[col for col in columns_to_drop if col in numeric_cols.columns])

# Nombre de colonnes par ligne
cols_per_row = 4

# Calcul du nombre de lignes nécessaires
num_rows = (len(numeric_df.columns) + cols_per_row - 1) // cols_per_row

# Création des boxplots
fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(16, 4 * num_rows))

# Si les axes sont en 1D (dans le cas où il y a juste une ligne)
if num_rows == 1:
    axes = [axes]

# Itération sur les colonnes et les axes pour créer les boxplots
for i, col in enumerate(numeric_df.columns):
    row = i // cols_per_row
    col_position = i % cols_per_row
    axes[row][col_position].boxplot(numeric_df[col], vert=False)
    axes[row][col_position].set_title(f'Boxplot of {col}')
    axes[row][col_position].set_xlabel(col)

# Supprimer les sous-graphiques vides
for i in range(len(numeric_df.columns), num_rows * cols_per_row):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

Les boxplots présentés peut se concentrer sur plusieurs aspects, notamment la détection des outliers, la dispersion des données, et les tendances générales.

### 1. **Number of views (7m) :**
Ce boxplot montre une distribution fortement asymétrique avec de nombreux outliers. La majorité des données sont concentrées près de la médiane, mais il existe des clients qui ont un nombre de vues bien supérieur à la moyenne.

La plupart des clients effectuent un nombre relativement faible de vues, mais un petit groupe d'clients est extrêmement actif, ce qui indique des super-clients ou des clients très engagés.

### 2. **Number of carts (7m) :**
Une distribution similaire à celle des vues avec un nombre important d'outliers. La plupart des clients ajoutent peu de produits au panier, tandis que quelques clients ont ajouté de nombreux produits.

Il pourrait être intéressant d'analyser plus en détail ces outliers pour comprendre s'il s'agit des clients très engagés ou d'un comportement spécifique à certains segments.

### 3. **Number of sessions (7m) :**
Distribution asymétrique avec plusieurs outliers, similaire aux deux premières métriques.

Quelques clients génèrent un nombre élevé de sessions, ce qui pourrait être corrélé à une forte utilisation ou à une navigation intensive.

### 4. **Count of products (7m) :**

## 8 - **Les histogrammes (représentation des proportions)**

```python


# Transformation des variables continues en catégories
df['avg_price_category'] = pd.cut(df['avg_price_7m'], bins=[0, 50, 100, 150, 200, df['avg_price_7m'].max()], labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

days_max = df['days_since_last_purchase'].max()
df['days_since_last_purchase_category'] = pd.cut(
    df['days_since_last_purchase'], 
    bins=[0, 7, 30, 60, days_max + 1],  # Ajout de +1 pour éviter la duplication
    labels=['<1 Week', '1 Week-1 Month', '1-2 Months', '>2 Months'],
    duplicates='drop'  # Supprimer les duplications si elles se produisent
)

# Liste des variables catégorielles à analyser, incluant les nouvelles catégories créées
categorical_vars = [ 
    'most_active_day', 
    'most_active_time', 
    'days_since_last_purchase_category'
]

# Nombre de colonnes par ligne
cols_per_row = 3

# Calcul du nombre de lignes nécessaires
num_rows = (len(categorical_vars) + cols_per_row - 1) // cols_per_row

# Création des subplots
fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(18, 6 * num_rows))

# Si les axes sont en 1D (dans le cas où il y a juste une ligne)
axes = axes.flatten()  # Rendre l'accès aux axes plus facile

# Itération sur les variables et les axes pour créer les graphiques
for i, var in enumerate(categorical_vars):
    # Calcul des proportions
    proportions = df[var].value_counts(normalize=True)
    
    # Affichage des proportions sous forme de graphique en barres avec couleur orange
    proportions.plot(kind='bar', ax=axes[i], color='orange')
    axes[i].set_title(f'Proportions of {var}')
    axes[i].set_ylabel('Proportion')
    axes[i].set_xlabel(var)
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=0)  # Échelle des abscisses horizontale

# Supprimer les sous-graphiques vides
for i in range(len(categorical_vars), num_rows * cols_per_row):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```


### Interprétation des Graphiques :

Ces trois histogrammes montrent les proportions de certaines variables catégorielles, fournissant une vue d'ensemble sur les comportements et préférences des clients.

1. **Proportions of most_active_day :**
   ici, nous montrons la répartition des jours de la semaine pendant lesquels les clients sont les plus actifs.
   La proportion de chaque jour est très proche, indiquant une activité des clients assez équilibrée tout au long de la semaine. Il n'y a pas de jour nettement dominant, ce qui suggère que l'engagement des clients est relativement constant indépendamment du jour. Cela pourrait indiquer que les clients interagissent avec la plateforme de manière régulière sans dépendance particulière à un jour précis.


2. **Proportions of most_active_time :**
   Cet histogramme illustre la répartition des moments de la journée (matin, après-midi, soirée, nuit) où les clients sont les plus actifs.
   Le matin est de loin le moment le plus actif, suivi de l'après-midi. L'activité diminue fortement en soirée et la nuit. Cela suggère que les clients préfèrent interagir avec la plateforme principalement durant la matinée et, dans une moindre mesure, l'après-midi. Ce comportement pourrait être lié à des habitudes de consommation comme la vérification de nouvelles offres ou la planification d'achats en début de journée.
   Pour maximiser l'engagement, il serait stratégique de concentrer les campagnes publicitaires, notifications, et offres spéciales principalement le matin et en début d'après-midi.
