# 4. Analyse des corrélations

Matrice de corrélation pour voir comment les variables numériques sont liées entre elles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'df' soit votre DataFrame contenant les données

# Retirer temporairement la colonne qualitative "most_active_day" pour l'analyse
df_numeric = df.drop(columns=["most_active_day", "user_id"])  # Retirer également "user_id" car ce n'est pas pertinent pour la corrélation

# Sélection des colonnes numériques uniquement
numeric_df = df_numeric.select_dtypes(include=[float, int])

# Calcul de la matrice de corrélation
correlation_matrix = numeric_df.corr()

# Tracer la heatmap de la matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de Corrélation des Variables Numériques')
plt.show()

# Réinsertion de "most_active_day" dans df après l'analyse si nécessaire
# df["most_active_day"] = df_original["most_active_day"]  # Si df_original est la version complète


```



### Points Clés de la Matrice de Corrélation :

1. **Corrélations Fortes (proches de 1 ou -1) :**
   - **`number_of_views_2m`, `number_of_views_5m`, `number_of_views_7m`** : Ces variables sont fortement corrélées entre elles, ce qui est attendu car elles représentent probablement le même comportement (le nombre de vues) sur différentes périodes de temps.
   - **`number_of_sessions_2m`, `number_of_sessions_5m`, `number_of_sessions_7m`** : De même, le nombre de sessions sur différentes périodes de temps est également fortement corrélé, suggérant une cohérence dans l'engagement des clients au fil du temps.
   - **`count_products_2m`, `count_products_5m`, `count_products_7m`** : Les corrélations entre ces variables sont également fortes, indiquant que les clients qui consultent plus de produits sur une période sont susceptibles de le faire également sur d'autres périodes.
   - **`total_purchase_value` et `count_products`** : Il existe une forte corrélation entre la valeur totale des achats et le nombre de produits consultés, ce qui indique que les clients qui voient plus de produits ont tendance à dépenser plus.

2. **Corrélations Modérées :**
   - **`number_of_carts` et `number_of_sessions/views`** : Le nombre de paniers (`carts`) est modérément corrélé avec le nombre de sessions et de vues, ce qui est logique, car plus d'interactions devraient conduire à plus de produits ajoutés au panier.
   - **`cart_abandonments` et `number_of_sessions/views`** : Il y a une corrélation modérée entre les abandons de paniers et les autres comportements d'engagement, ce qui pourrait indiquer que les clients plus actifs sont aussi ceux qui abandonnent plus souvent leurs paniers.

3. **Corrélations Faibles à Négatives :**
   - **`brand_loyalty`** : La loyauté à la marque montre généralement des corrélations faibles ou négatives avec la plupart des autres variables, suggérant que la fidélité à une marque spécifique ne dépend pas fortement du nombre de sessions, de vues ou même des achats totaux.
   - **`days_since_last_purchase`** : Cette variable est faiblement corrélée avec les autres, sauf peut-être avec des variables indiquant l'activité passée (comme les sessions et les vues), ce qui est attendu puisque plus il y a d'activité récente, plus le dernier achat est récent.
   - **`avg_price`** : Les prix moyens semblent avoir une corrélation assez faible avec la plupart des autres variables, indiquant que le montant que les clients sont prêts à dépenser n'est pas directement lié au nombre de vues ou de sessions.

### Interprétation Globale :

- **Comportements d'Engagement et d'Achat :** Les clients qui sont actifs sur la plateforme en termes de vues, de sessions et de produits consultés montrent une cohérence dans leur comportement, avec des corrélations fortes entre ces mesures. Cela indique que les comportements d'engagement sont souvent interconnectés.
  
- **Valeur des Achats :** Les corrélations montrent que l'engagement (vues, sessions, produits consultés) est lié positivement à la valeur totale des achats. Plus un utilisateur est actif, plus il est probable qu'il dépense davantage.

- **Abandons de Panier et Loyauté :** Les abandons de panier sont liés à l'activité globale, mais la loyauté à la marque est une variable plus indépendante, peu corrélée avec les autres indicateurs d'engagement.

## 6. Analyse des outliers

```python
# Liste des colonnes à supprimer
columns_to_drop = ['number_of_views_2m', 'number_of_views_5m', 'number_of_carts_2m',
                   'number_of_carts_5m', 'number_of_sessions_2m', 'number_of_sessions_5m',
                   'count_products_2m', 'count_products_5m', 'avg_price_2m', 'avg_price_5m']

# Suppression des colonnes si elles existent dans le DataFrame
numeric_df = numeric_df.drop(columns=[col for col in columns_to_drop if col in numeric_df.columns])

# Nombre de colonnes par ligne
cols_per_row = 4

# Calcul du nombre de lignes nécessaires
num_rows = (len(numeric_df.columns) + cols_per_row - 1) // cols_per_row

# Création des boxplots
fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(16, 4 * num_rows))

# Si les axes sont en 1D (dans le cas où il y a juste une ligne)
if num_rows == 1:
    axes = [axes]

# Itération sur les colonnes et les axes pour créer les boxplots
for i, col in enumerate(numeric_df.columns):
    row = i // cols_per_row
    col_position = i % cols_per_row
    axes[row][col_position].boxplot(numeric_df[col], vert=False)
    axes[row][col_position].set_title(f'Boxplot of {col}')
    axes[row][col_position].set_xlabel(col)

# Supprimer les sous-graphiques vides
for i in range(len(numeric_df.columns), num_rows * cols_per_row):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

Les boxplots présentés peut se concentrer sur plusieurs aspects, notamment la détection des outliers, la dispersion des données, et les tendances générales.

```python

# import numpy as np

# Lire le fichier Parquet
user_stats_df_pandas = pd.read_parquet("/home/jovyan/work/user_stats_df_output.parquet")

# Séparer les colonnes qualitatives et quantitatives
qualitative_columns = ['preferred_brand', 'preferred_category', 'most_active_day', 'most_active_time', 'user_segment']
quantitative_columns = ['number_of_views_7m',
                        'number_of_carts_7m',
                        'number_of_sessions_7m',
                        'count_products_7m',
                        'avg_price_7m',
                        'total_purchase_value',
                        'days_since_last_purchase',
                        'cart_abandonments', 
                        'brand_loyalty', 
                        'avg_session_duration'
]

# Préparer les transformateurs pour les colonnes qualitatives et quantitatives
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), quantitative_columns),
        ('cat', OneHotEncoder(), qualitative_columns)
    ]
)

# Appliquer les transformations aux données
user_stats_processed = preprocessor.fit_transform(user_stats_df_pandas)

# Effectuer l'ACP
pca = PCA(n_components=3)  # Réduire à 3 composantes principales
principal_components = pca.fit_transform(user_stats_processed)

# Convertir les résultats de l'ACP en DataFrame Pandas
principal_df = pd.DataFrame(data=principal_components, columns=['principal_component_1', 'principal_component_2', 'principal_component_3'])

# Ajouter les labels clients au DataFrame des composantes principales
final_df = pd.concat([principal_df, user_stats_df_pandas[['user_id']]], axis=1)

# Fonction pour tracer le cercle des corrélations avec uniquement les vecteurs ayant une forte contribution
def plot_correlation_circle(pcs, n_comp, pca, axis_ranks, labels=None, label_rotation=0, lims=None, threshold=0.1):
    for d1, d2 in axis_ranks:
        if d2 < n_comp:
            fig, ax = plt.subplots(figsize=(12, 12))  # Taille augmentée
            
            # Cercle
            ax.plot([-1, 1], [0, 0], color='grey', ls='--')
            ax.plot([0, 0], [-1, 1], color='grey', ls='--')
            circle = plt.Circle((0, 0), 1, facecolor='none', edgecolor='b')
            ax.add_artist(circle)

            # Variables
            vectors = pcs[:, [d1, d2]]
            for i, (x, y) in enumerate(vectors):
                plt.plot([0, x], [0, y], 'k-', alpha=0.5)  # Tracer les vecteurs
                # Afficher uniquement les étiquettes significatives
                if x**2 + y**2 > threshold:  # Seulement si la contribution est significative
                    if labels is None:
                        plt.text(x, y, f"Var{i+1}", fontsize=12, ha='center', va='center', rotation=label_rotation, color="blue", alpha=0.75)
                    else:
                        plt.text(x, y, labels[i], fontsize=12, ha='center', va='center', rotation=label_rotation, color="blue", alpha=0.75)
            
            # Limites
            if lims is not None:
                plt.xlim(lims)
                plt.ylim(lims)
            else:
                plt.xlim(-1, 1)
                plt.ylim(-1, 1)

            plt.xlabel(f'F{d1+1} ({round(pca.explained_variance_ratio_[d1]*100, 1)}%)')
            plt.ylabel(f'F{d2+1} ({round(pca.explained_variance_ratio_[d2]*100, 1)}%)')
            plt.title(f"Cercle des corrélations F{d1+1} et F{d2+1}")
            plt.show()

# Calcul des composantes principales des variables
pcs = pca.components_.T

# Labels des variables
labels = quantitative_columns + list(preprocessor.named_transformers_['cat'].get_feature_names_out())

# Plot du cercle des corrélations pour F1 et F2, et pour F1 et F3 avec un seuil ajusté pour la contribution
plot_correlation_circle(pcs, n_comp=3, pca=pca, axis_ranks=[(0, 1), (0, 2)], labels=labels, threshold=0.1)
```
